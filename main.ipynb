```python
import pandas as pd

df = pd.read_csv('./euro2024_players.csv')
```

```python
# Exibir as primeiras linhas do dataset
print(df.head())

# Obter informações sobre o dataset
print(df.info())

# Estatísticas descritivas das colunas numéricas
print(df.describe())
```

```python
df['Goals'] = pd.to_numeric(df['Goals'], errors='coerce')
```

```python
top_scorers = df.sort_values(by='Goals', ascending=False).head(10)


print(top_scorers[['Name', 'Goals']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


dark_palette = sns.color_palette("colorblind", 1)

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Goals', y='Name', data=top_scorers, palette=dark_palette)
plt.title('Top 10 Jogadores com Mais Gols')
plt.xlabel('Gols')
plt.ylabel('Jogadores')
plt.show()
```

```python
# Converter a coluna 'Goals' para numérica, se necessário
df['Goals'] = pd.to_numeric(df['Goals'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
df['MarketValue'] = pd.to_numeric(df['MarketValue'], errors='coerce')

# Selecionar apenas as colunas numéricas para a análise de correlação
numeric_columns = ['Goals', 'Age', 'MarketValue', 'Height']

# Calcular a matriz de correlação
corr_matrix = df[numeric_columns].corr()

# Plotar a matriz de correlação usando Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação entre Variáveis Numéricas')
plt.show()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Goals'] = pd.to_numeric(df['Goals'], errors='coerce')

# Plotar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Goals'], alpha=0.5, color='b')
plt.title('Idade vs Gols')
plt.xlabel('Idade')
plt.ylabel('Gols')
plt.grid(True)
plt.show()
```
